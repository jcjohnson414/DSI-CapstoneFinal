```python
ihg1 = pd.read_csv('ihg_hotels_descriptions2.csv')
ihg2 = pd.read_csv('ihg_hotels_descriptions3.csv')
ihg3 = pd.read_csv('ihg_hotels_descriptions4.csv')
ihg4 = pd.read_csv('ihg_hotels_descriptions5.csv')
ihg5 = pd.read_csv('ihg_hotels_descriptions6.csv')
missing = pd.read_csv('missing_ihg_hotels3.csv')
```

```python
ihg1 = ihg1.drop_duplicates(subset=['url'])
ihg2 = ihg2.drop_duplicates(subset=['url'])
ihg3 = ihg3.drop_duplicates(subset=['street'])
ihg4 = ihg4.drop_duplicates(subset=['url'])
ihg5 = ihg5.drop_duplicates(subset=['url'])
```

```python
merged = pd.concat([ihg1, ihg2, ihg3, ihg4, ihg5], ignore_index=True)
```

```python
merged.shape
```

```python
merged.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
merged.shape
```

```python
merged['brand'] = ''
```

```python
for index, row in merged.iterrows():
    if 'InterContinental' in row['name']:
        merged['brand'][index] = 'InterContinental'
    elif 'Indigo' in row['name']:
        merged['brand'][index] = 'Indigo'
    elif 'Kimpton' in row['name']:
        merged['brand'][index] = 'Kimpton'
    elif 'Crowne Plaza' in row['name']:
        merged['brand'][index] = 'Crowne Plaza'
    elif 'EVEN' in row['name']:
        merged['brand'][index] = 'EVEN'
    elif 'Staybridge' in row['name']:
        merged['brand'][index] = 'Staybridge'
    elif 'Express' in row['name']:
        merged['brand'][index] = 'Holiday Inn Express'
    elif 'Avid' in row['name']:
        merged['brand'][index] = 'Avid'
    elif 'Candlewood' in row['name']:
        merged['brand'][index] = 'Candlewood'
#     elif 'Waldorf' in row['name']:
#         hilton['brand'][index] = 'Waldorf Astoria'
#     elif 'Canopy' in row['name']:
#         hilton['brand'][index] = 'Canopy'
#     elif 'Curio' in row['name']:
#         hilton['brand'][index] = 'Curio'
#     elif 'Hilton Grand' in row['name']:
#         hilton['brand'][index] = 'Hilton Grand Vacations'
#     elif 'Tapestry' in row['name']:
#         hilton['brand'][index] = 'Tapestry'
#     elif 'Motto' in row['name']:
#         hilton['brand'][index] = 'Motto'
    else:
        merged['brand'][index] = 'Holiday Inn'
```
