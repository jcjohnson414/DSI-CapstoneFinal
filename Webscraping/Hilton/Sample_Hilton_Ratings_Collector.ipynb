```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium.webdriver import Chrome
# driver = Chrome(executable_path='/Users/chrisjohnson/Downloads/chromedriver')
import time
import numpy as np
import json
from datetime import datetime

pd.set_option('max_columns', None)
pd.set_option('max_rows', None)
```

```python
def get_amenities(state):
    dataframe = pd.read_csv('../'+ state)
    counter=0
    ratings = []
    missing = []
    for index, row in dataframe.iterrows():
        name = row['name']
        url = row['url']
        desc = row['description']
        street = row['streetAddress']
        locality = row['addressLocality']
        state = row['addressRegion']
        postalcode = row['postalCode']
        latitude = row['latitude']
        longitude = row['longitude']
        try:
            url2 = url
#             print(url)
            res = requests.get(url2)
#             driver.get(url)
#             time.sleep(5)
            hotel_soup = BeautifulSoup(res.content, 'lxml')
            
            test = hotel_soup.find('script', attrs = {'id':'__NEXT_DATA__'})
            ratingdict ={}
            test_content = test.contents[0]
            rating = test_content[test_content.find('"rating"')+9:test_content.find('"rating"')+14]
            ratingdict['name'] = name
            ratingdict['rating'] = rating.split(',')[0]
            ratings.append(ratingdict)
            
        except:
            rating = {}
            rating['name'] = name
            rating['rating'] = 'N/A'
            ratings.append(rating)      

#         print("="*50)
        if (counter + 1) % 1 == 0:
            print(f"Attempted to collect {counter} hotels amenities list at {datetime.now()}.")
        counter += 1
    ratings_df = pd.DataFrame(ratings)
#     missing_df = pd.DataFrame(missing)
#     mergedratings = pd.concat([dataframe, amenities2_df], axis=0, ignore_index=True)
    ratingsmerged = pd.merge(dataframe, ratings_df, on='name')
#     missingmerged = pd.merge(dataframe, missing_df, on='name')
    return ratingsmerged # missingmerged
```

```python
%%time
california = get_amenities('california_descriptions_2.csv')
```
